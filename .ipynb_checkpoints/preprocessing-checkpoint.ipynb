# Preprocess the video data

```python
import cv2

# capture video and read frames
vidcap = cv2.VideoCapture('data/game/MafiaVideogame.mp4')
success, image = vidcap.read()
count = 0

while success:
    
    # save frames as images
    if count >= 7110:
        cv2.imwrite("data/game/frames/frame_%d.jpg" % count, image)
        print('Read a new frame:', count)
        count += 90
        vidcap.set(1, count)
    
    success, image = vidcap.read()
    count += 1
```
