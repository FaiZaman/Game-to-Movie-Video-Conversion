# Preprocess the video data

```python
import cv2
```

```python
# capture video and read frames
vidcap = cv2.VideoCapture('data/game/MafiaVideogame.mp4')
frame_number = 7110
vidcap.set(1, frame_number)
success, image = vidcap.read()

while success and frame_number < 232110:
    
    # save frames as images
    cv2.imwrite("data/game/frames/frame_%d.jpg" % frame_number, image)
    print('Read a new frame:', frame_number)
    frame_number += 450
    
    vidcap.set(1, frame_number)
    success, image = vidcap.read()
```
